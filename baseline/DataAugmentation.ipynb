```python
from PIL import Image
from PIL import ImageEnhance as ie
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img = Image.open("../data/train-jpg/train_30000.jpg")
```

```python
img = img.convert('RGB')
```

```python
plt.imshow(img)
```

```python
images = []

images.append(ie.Color(img).enhance(0.6))
images.append(ie.Color(img).enhance(1.4))

images.append(ie.Brightness(img).enhance(0.6))
images.append(ie.Brightness(img).enhance(1.4))

images.append(ie.Contrast(img).enhance(0.6))
images.append(ie.Contrast(img).enhance(1.4))

images.append(ie.Sharpness(img).enhance(0.6))
images.append(ie.Sharpness(img).enhance(1.4))
```

```python
for im in images:
    plt.figure()
    plt.imshow(im)
```
